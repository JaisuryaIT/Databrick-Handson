```python
from pyspark.sql.functions import concat_ws

df = spark.table("workspace.default.bronze_customers")

silver_df = df.filter("email_address IS NOT NULL") \
              .withColumn(
                  "full_name",
                  concat_ws(" ", df.first_name, df.last_name)
              )

display(silver_df)

silver_df.write.mode("overwrite").saveAsTable("workspace.default.silver_customers")
```
